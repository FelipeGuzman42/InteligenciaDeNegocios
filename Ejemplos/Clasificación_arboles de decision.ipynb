### 2. Limpieza y preparacin de los datos

Primero vamos a ejecutar los pasos de limpieza de los datos, relacionados el tratamiento de ausencias y registros duplicados.

```python
# Es recomendable que todos los pasos de limpieza y preparacin se realicen sobre otro archivo.
df_tracks_t = df_tracks
# Primero eliminaremos aquellas variables con muchas ausencias.
df_tracks_t = df_tracks_t.drop(['decibel_range', 'second_artist'], axis=1)
```

```python
# Eliminacin registros con ausencias
df_tracks_t = df_tracks_t.dropna()
# Eliminacin de registros duplicados.
df_tracks_t = df_tracks_t.drop_duplicates()
```

Es muy importante analizar el objetivo del modelo analtico a construir antes de iniciar a tomar decisiones. En este caso, por ejemplo, se debe aplicar la regla dada por el negocio, sobre qu significa una cancin popular y que no, para generar la nueva variable que ser la clase (variable objetivo).

Esta nueva columna la derivaremos a partir de la variable "popularity" utilizando un umbral que nos permita determinar cundo una cancin es popular. Si el valor en esta variable es mayor que el umbral entonces ese registro se clasifica como "popular" y ser indicado con un "1", de lo contrario ser "no popular y esta categora la representaremos con un "0".

```python
# Podemos ver los estadsticos de la variable "popularity"
df_tracks_t['popularity'].describe()
```

```python
# Ahora definimos la funcin que nos va a permitir construir nuestra clase.
def label_popularity (row):
    if row['popularity'] > 33 :
      return 1
    return 0
df_tracks_t['popularity_label']=df_tracks_t.apply (lambda row: label_popularity(row), axis=1)    
```

```python
df_tracks_t.head()
```

```python
# Cantidad de datos de cada clase
pd.value_counts(df_tracks_t['popularity_label'])
```

```python
# Eliminaremos las variables que consideramos no son tiles para la tarea que queremos resolver, como los identificadores.
df_tracks_t = df_tracks_t.drop(['id', 'id_artists'], axis=1)
```

Un aspecto muy importante para tener en cuenta son los requerimientos de entrada de los algoritmos de aprendizaje implementados en las libreras utilizadas. En particular, scikit-learn requiere que todos los atributos sean numricos y que la variable objetivo, en una tarea de clasificacin, est codificada con nmeros.

```python
# Vamos a seleccionar de nuestro conjunto solo los atributos numricos.
number_cols = df_tracks_t.dtypes[(df_tracks_t.dtypes == np.int64) | (df_tracks_t.dtypes == np.float64)].index 
number_cols = df_tracks_t.select_dtypes(include = ['int64','float']).columns
number_cols
```

### 3. Construccin del modelo

Los algoritmos supervisados implementados en scikit-learn requieren que las variables de entrada estn separadas de la variable objetivo. 

```python
# Se selecciona la variable objetivo, en este caso "Popularity_label".
Y = df_tracks_m['popularity_label']
# Del conjunto de datos se elimina la variable "Popularity_label"
X = df_tracks_m.drop(['popularity_label'], axis=1)
```

```python
# Dividir los datos en entrenamiento y test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
# Crear el objeto de arbol de decisin. Utilicemos como criterio de pureza la entropa.
arbol = DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```python
# Entrenar el modelo de arbol de decisin con los datos de entrenamiento
arbol = arbol.fit(X_train,Y_train)
```

Para determinar el rendimiento del modelo podemos en primer lugar, generar la matriz de confusin.

```python
# Determinamos las predicciones del modelo sobre el conjunto test.
y_pred = arbol.predict(X_test)
```

```python
# Se genera la matriz de confusin
confusion_matrix(Y_test, y_pred)
```

```python
# Se puede visualizar la matriz de confusin
plot_confusion_matrix(arbol, X_test, Y_test)  
plt.show()  
```

Tambin podemos generar diferentes mtricas de rendimiento como la excatitud, el recall, la precisin y la puntuacin F1. 

```python
print('Exactitud: %.2f' % accuracy_score(Y_test, y_pred))
print("Recall: {}".format(recall_score(Y_test,y_pred)))
print("Precisin: {}".format(precision_score(Y_test,y_pred)))
print("Puntuacin F1: {}".format(f1_score(Y_test,y_pred)))
```

Para tener una panorama completo podemos generar un reporte.

```python
# Mostrar reporte de clasificacin
print(classification_report(Y_test, y_pred))
```

Cuando tenemos una prediccin exacta, hay que revisar el modelo porque indica un posible error. ### 5. Construccin del modelos con bsqueda de hiperparmetros.
En el caso de los rboles de decisin los hiperparmetros principales que hay que ajustar estn relacionados con aquellos que controlan la complejidad, pero tambin el criterio o medida de pureza puede afectar su rendimiento. En general, estos son:  

* **criterion** : {gini, entropy}, predeterminado = "gini".

La funcin para medir la calidad de una divisin. Los criterios admitidos son "gini" y "entropa".

* **max_depth:** int, predeterminado=None.

La profundidad mxima del rbol. Si es None, los nodos se expanden hasta que todas las hojas sean puras o hasta que todas las hojas contengan menos de min_samples_split muestras.

* **min_samples_split** : int or float, predeterminado=2.

El nmero mnimo de muestras necesarias para dividir un nodo interno



Cmo saber los mejores valores de los hiperparmetro de un algoritmo? Estos hay que ajustarlo a los datos y no podemos utilizar el conjunto test para realizar esta tarea. Debemos entonces recurrir a las tcnicas de validacin las cuales, en general, puedes implementarlas de dos maneras: utilizando un conjunto de validacin o empleando la validacin cruzada de k particiones (k-fold cross validation).

Veamos cmo hacerlo.

```python
# Dividimos el conjunto de entrenamiento en dos: una para la construccin del modelo (sera el nuevo conjunto de 
# entrenamiento) y otro para la validacin, el cul ser utilizado para determinar el rendimiento del modelo con una
# combinacin especfica de hiperparmetros.
X_trainval, X_val, Y_trainval, Y_val = train_test_split(X_train, Y_train, test_size = 0.15, random_state = 0)
```

Ahora utilicemos la otra opcin.

b) Validacin cruzada de k-particiones. En este caso el conjunto de entrenamiento se divide en k particiones disjuntas. Luego, se fija un valor de hiperparmetro y se toma como conjunto de validacin la primera particin y el resto para entrenar el modelo. Este proceso se repite para cada particin. Al finalizar, se determina el valor promedio de las mtricas seleccionadas. Este ciclo se repite sobre diferentes valores del hiperparmetro y se seleccionan aquel que ofrezcan el mejor rendimiento. 

Scikit-learn ofrece algunos mtodos que automatizan el proceso de buscar los valores de los hiperparmetros. Uno de ellos es GridSearchCV, el cual se basa en la validacin cruzada de k-particiones. Veamos cmo utilizarlo.

```python
# Fijemos el nmero de particiones. Utilizaremos K = 10.
particiones = KFold(n_splits=10, shuffle=True, random_state = 0)
```

```python
# Establecemos el espacio de bsqueda para los hiperparmetros que deseamos ajustar. 
param_grid = {'criterion':['gini', 'entropy'],'max_depth':[4,6,8,10,20],'min_samples_split':[2, 3, 4, 5]}
```

```python
# Definimos el modelo sin ningn valor de estos hiperparmetros
arbol = DecisionTreeClassifier(random_state=0)
```

```python
# Ahora utilizamos GridSearch sobre el grid definido y con 10 particiones en la validacin cruzada.
mejor_modelo = GridSearchCV(arbol, param_grid, cv=particiones)
# Ajuste del modelo
mejor_modelo.fit(X_train, Y_train)
```

```python
# Podemos ver cul fue el resultado de la bsqueda (mejores valores de hiperparmetros)
mejor_modelo.best_params_
```

```python
# Obtener el mejor modelo.
arbol_final = mejor_modelo.best_estimator_
# Probemos ahora este modelo sobre test.
y_pred_train = arbol_final.predict(X_train)
y_pred_test = arbol_final.predict(X_test)
print('Exactitud sobre entrenamiento: %.2f' % accuracy_score(Y_train, y_pred_train))
print('Exactitud sobre test: %.2f' % accuracy_score(Y_test, y_pred_test))
```

```python
print(classification_report(Y_test, y_pred_test))
```
